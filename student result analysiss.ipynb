```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("student_scores.csv")
print(df.head())
```

Drop Unnamed Column

```python
df = df.drop("Unnamed: 0", axis = 1)
print(df.head())
```

```python
df.head()
```

gender distribution

```python
plt.figure(figsize = (5,5))
ax=sns.countplot(data = df, x = "Gender")
plt.title("Gender Distibution")
ax.bar_label(ax.containers[0])
plt.show()
```

# From the above chart we have analysed that : 
# the number of females in the data is more than that of the males

```python
gb = df.groupby("ParentEduc").agg({"MathScore":'mean', "ReadingScore":'mean', "WritingScore":'mean'})
print(gb)
```

```python
plt.figure(figsize = (5,5))
sns.heatmap(gb, annot = True)
plt.title("Relationship between Parents Education and Students Score")
plt.show()
```

```python
#from the above chart we have concluded that parents education have great impact on students marks
```

```python
gb1 = df.groupby("ParentMaritalStatus").agg({"MathScore":'mean', "ReadingScore":'mean', "WritingScore":'mean'})
print(gb1)
```

```python
plt.figure(figsize = (4,4))
sns.heatmap(gb1, annot = True)
plt.title("Relationship between Parents Marital Status and Students Score")
plt.show()
```

# from the above chart we have concluded that there is no impact of parents marital status on students results

Distribution of ethnic group

```python
l = ["group A", "group B", "group C", "group D", "group E" ]
groupA = df.loc[(df['EthnicGroup' ]== 'group A')].count()
groupB = df.loc[(df['EthnicGroup' ]== 'group B')].count()
groupC = df.loc[(df['EthnicGroup' ]== 'group C')].count()
groupD = df.loc[(df['EthnicGroup' ]== 'group D')].count()
groupE = df.loc[(df['EthnicGroup' ]== 'group E')].count()

mlist = [groupA["EthnicGroup"],groupB["EthnicGroup"],groupC["EthnicGroup"],groupD["EthnicGroup"],groupE["EthnicGroup"]]
plt.pie(mlist, labels = l, autopct = "%1.2f%%")
plt.title("Distribution of Ethnic Groups")
plt.show()
```
